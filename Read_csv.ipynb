```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

def convert_to_numbers(x):
    if isinstance(x, str):
        x = x.strip().upper()
        if x.endswith("K"):
            return float(x[:-1]) * 1000
    return pd.to_numeric(x, errors='coerce')

data = pd.read_csv("Slurm-data.csv", on_bad_lines='skip', header=None)
data.columns = ["JobID", "UID", "ReqNodes", "TimeLimitRaw", "ElapsedRaw"]

data["ReqNodes"] = data["ReqNodes"].apply(convert_to_numbers)

data_cleaned = data.dropna(subset=["ReqNodes", "ElapsedRaw"])

figure = px.scatter(data_cleaned, x="ReqNodes", y="ElapsedRaw", title="Elapsed Time vs Requested Nodes",labels={"ReqNodes": "Requested Nodes", "ElapsedRaw": "Elapsed Time (seconds)"})

figure.show()



```
